```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_name='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/coursera/project/kc_house_data_NaN.csv'
data=pd.read_csv(file_name)
data
```

```python
data.drop("id", axis=1, inplace=True)
data.drop("Unnamed: 0", axis=1, inplace=True)
data.describe()
```

```python
data["price"]
```

# Affiche la somme des prix des maison en fonction du nombre de chambre

```python
data.groupby("price").bedrooms.sum()
```

# Ressort la somme et le minimum des valeur du dataset

```python
data.agg(['sum', 'min'])
```

# Donne les  maisons les plus cher du dataset

```python
data.sort_values(by='price')[-5:]
```

# Compte le nombre de maison par nombre d'étage 

```python
data['floors'].value_counts().to_frame()
```

# Moyenne des prix en fonction du nombre de chambre 

```python
data.groupby("bedrooms").price.mean()
```

# Affiche les prix en fonction du nombre de salles de bain pour les maison ayant 3 chambre

```python
data[(data.bedrooms==3)].groupby("price").bathrooms.sum().plot()
```

# Affiche le prix en moyenne pour le nombre de chambre donné

```python
total_chambre=data.pivot_table('price',index='bedrooms',aggfunc=np.mean)
total_chambre.tail()
```

```python
total_chambre.plot(title="total chambre par prix")
```

# Affiche si le prix est negativement ou positivement correlé avec le sqft_above (mètre carré)

```python
sns.regplot(x="sqft_above", y="price", data=data)
plt.ylim(0,)
```
